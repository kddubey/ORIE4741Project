```julia
using LowRankModels, Random, LinearAlgebra, Plots, Statistics, CSV
```

```julia
Y = CSV.read("X_tr.csv", header=false);
```

```julia
size(Y)
```

```julia
Y = convert(Matrix, Y)
Ω = findall(.!ismissing.(Y) .& .!isnan.(Y));
```

```julia
# what's there to impute
length(Ω)/length(Y)
```

```julia
losses = vcat(Loss[HingeLoss() for _ in 1:13], Loss[L1Loss(), HingeLoss()])
size(losses)
```

```julia
Y = convert(Array{Any,2}, Y)
typeof(Y[1,1])
```

```julia
Y[:,1:13] = Int.(Y[:,1:13])
Y[:,15] = Int.(Y[:,15])
typeof(Y[1,1])
```

```julia
reg = ZeroReg()
k = 5
glrm = GLRM(Y, losses, reg, reg, k, obs=Ω);
```

```julia
X,W,ch = fit!(glrm); 
```
