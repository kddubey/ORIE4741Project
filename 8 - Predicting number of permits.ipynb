# Create counts of permits by zipcode

Do as a proportion of the whole dataset so that it's not influenced by the dataset size. 

```python
zipcode_list = zipcode_data['Zipcode'].values
zipcode_list
```

```python
permits_by_zipcode_train = [np.sum(data_train['Zipcode'] == z) for z in zipcode_list]
permits_by_zipcode_train /= sum(permits_by_zipcode_train)
permits_by_zipcode_train
```

```python
permits_by_zipcode_test = [np.sum(data_test['Zipcode'] == z) for z in zipcode_list]
permits_by_zipcode_test /= sum(permits_by_zipcode_test)
```

### Is the number of permits in a zipcode alone a good predictor?

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(permits_by_zipcode_test, permits_by_zipcode_train))
print('Mean Squared Error:', metrics.mean_squared_error(permits_by_zipcode_test, permits_by_zipcode_train))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(permits_by_zipcode_test, permits_by_zipcode_train)))
```

Answer: yes. The training and test data are split randomly, and the proportions should be approximately equal.

# Use subset of zipcodes to predict rest of zipcodes

Will be a "small data" problem: only 27 zipcodes

```python
permits_by_zipcode_all = [(np.sum(data_train['Zipcode'] == z) + np.sum(data_test['Zipcode'] == z)) for z in zipcode_list]
```

```python
zipcode_train, zipcode_test, permits_train, permits_test = train_test_split(zipcode_data, permits_by_zipcode_all, test_size=0.2, random_state=10)
```

```python
# simple regression
y_train = permits_train
nrows_train = len(y_train)

X_train = zipcode_train.drop(columns=['Zipcode'])

regressor = LinearRegression()
regressor.fit(X_train, y_train)

y_test = permits_test
nrows_test = len(y_test)

X_test = zipcode_test.drop(columns=['Zipcode'])

y_pred_train = regressor.predict(X_train)
y_pred = regressor.predict(X_test)

plt.scatter(y_test, y_pred,  color='gray')
plt.show()

print('Mean Absolute Error on training data:', metrics.mean_absolute_error(y_train, y_pred_train))
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

print('Regressor coefficients: ')
for i in range(len(regressor.coef_)):
    print(X_train.columns[i] + ": " + str(regressor.coef_[i]))

print('Regressor intercept: ', regressor.intercept_)

print('\nTrue value \t Predicted value')
for i in range(nrows_test):
    print(str(y_test[i]) + "\t\t" + str(y_pred[i]))
```

```python
selected_columns = ['Fire total', 'Business total', 'Restaurant total', 'Restaurant inspection score']

y_train = permits_train
nrows_train = len(y_train)

X_train = zipcode_train[selected_columns]

regressor = LinearRegression()
regressor.fit(X_train, y_train)

y_test = permits_test
nrows_test = len(y_test)

X_test = zipcode_test[selected_columns]

y_pred_train = regressor.predict(X_train)
y_pred = regressor.predict(X_test)

plt.scatter(y_test, y_pred,  color='gray')
plt.show()

print('Mean Absolute Error on training data:', metrics.mean_absolute_error(y_train, y_pred_train))
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

print('Regressor coefficients: ')
for i in range(len(regressor.coef_)):
    print(X_train.columns[i] + ": " + str(regressor.coef_[i]))

print('Regressor intercept: ', regressor.intercept_)

print('\nTrue value \t Predicted value')
for i in range(nrows_test):
    print(str(y_test[i]) + "\t\t" + str(y_pred[i]))
```

# Time series prediction

```python
year_beginnings = {} #inclusive
year_ends = {} #not inclusive

for i in range(2013, 2019):
    year_beginnings[i] = datetime.datetime(i, 1, 1)
    year_ends[i] = datetime.datetime(i+1, 1, 1)
    
print(year_beginnings)
print(year_ends)
```

```python
zipcode_data_by_year = {}

for i in range(2013, 2019):
    print('building-permit-applications-data/zipcode_data_year_' + str(i) + '.csv')
    zipcode_data_by_year[i] = pd.read_csv('building-permit-applications-data/zipcode_data_year_' + str(i) + '.csv')
```

```python
permits_by_zipcode_year = {}

for i in range(2013, 2019):
    print(i)
    min_of_year = year_beginnings[i]
    max_of_year = year_ends[i]
    
    permits_by_zipcode_year[i] = [np.sum((pd.to_datetime(data_train['Filed Date'], format='%m/%d/%Y') >= min_of_year) &\
(pd.to_datetime(data_train['Filed Date'], format='%m/%d/%Y') < max_of_year) & (data_train['Zipcode'] == z)) +\
np.sum((pd.to_datetime(data_test['Filed Date'], format='%m/%d/%Y') >= min_of_year) &\
(pd.to_datetime(data_test['Filed Date'], format='%m/%d/%Y') < max_of_year) & (data_test['Zipcode'] == z)) for z in zipcode_list]
```
